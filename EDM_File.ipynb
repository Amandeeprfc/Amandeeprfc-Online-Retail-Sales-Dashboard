### Nagative values issues has been resolved

```python
df.isna().sum()
```

```python
# Define a function to get the most frequent non-null description
def most_frequent(x):
    return x.dropna().mode().iloc[0] if not x.dropna().empty else np.nan

# Map most frequent description to NaNs
desc_map = df.groupby('StockCode')['Description'].transform(most_frequent)
df['Description'] = df['Description'].fillna(desc_map)
```

```python
df['Description'] = df.groupby('StockCode')['Description'].transform(
    lambda x: x.fillna(method='ffill').fillna(method='bfill')
)
```

```python
df[df['StockCode'].astype(str).str.contains('[A-Za-z]', regex=True)]
```

```python
df[pd.isna(df['Description'])]['StockCode'].value_counts(ascending=False)
```

```python
# Step 1: Count occurrences of each StockCode
stockcode_counts = df['StockCode'].value_counts()

# Step 2: Identify which StockCodes occur only once
single_occurrence = stockcode_counts[stockcode_counts == 1].index

# Step 3: Replace NaN descriptions with 'No Description' only for those
df.loc[
    (df['StockCode'].isin(single_occurrence)) & (df['Description'].isna()),'Description'] = 'No Description'
```

```python
df[pd.isna(df['Description'])]['StockCode'].value_counts(ascending=False)
```
